```python
class User:
    def __init__(self, name, balance=10000, correct_pin=1234):
        self.name = name
        self.balance = balance
        self.correct_pin = correct_pin

def atm_simulation(user):
    print(f'''
        ******************************************
        *        Welcome {user.name} to          *
        *      ATM Machine Simulator             *
        ******************************************
    ''')

    pin_attempts = 0

    while True:
        pin = int(input('Enter Your PIN: '))
        if pin != user.correct_pin:
            pin_attempts += 1
            if pin_attempts == 3:
                print("Too many incorrect attempts. Exiting...")
                return
            print("Incorrect PIN. Please try again.")
            continue
        else:
            break

    while True:
        print('''
            ******************************************
            *       Options you can Exercise are:   *
            *       1) Balance                      *
            *       2) Withdraw                     *
            *       3) Deposit                      *
            *       4) Change PIN                   *
            *       5) Exit                         *
            ******************************************
        ''')
        try:
            choose = int(input('Select Your Transaction from the above options: '))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue

        if choose == 1:
            print(f'Available Account Balance: ${user.balance:.2f}')
        elif choose == 2:
            withdraw_amount = float(input('Enter Withdrawal Amount: $'))
            if withdraw_amount <= user.balance:
                user.balance -= withdraw_amount
                print(f'Withdrawal successful. Your available balance is ${user.balance:.2f}')
            else:
                print('Insufficient Balance')
        elif choose == 3:
            deposit_amount = float(input('Enter Deposit Amount: $'))
            if deposit_amount > 0:
                user.balance += deposit_amount
                print(f'Deposit successful. New Balance is ${user.balance:.2f}')
            else:
                print('Invalid amount')
        elif choose == 4:
            new_pin = int(input("Enter the New PIN: "))
            user.correct_pin = new_pin
            print("PIN successfully changed.")
        elif choose == 5:
            print("Thank you for using the ATM. Have a nice day!")
            break
        else:
            print('Invalid choice. Please select a number between 1 and 5.')

# Example usage:
user = User("John Doe")
atm_simulation(user)
```
